# Solve `CartPole-v0` with `REINFORCE` as a policy-gradient method

```python
import gym
import matplotlib.pyplot as plt
import numpy as np
import torch
from torch.distributions import Categorical

from agent import Agent
from reinforce import reinforce, sample_trajectory
```

```python
env = gym.make('CartPole-v0')
```

```python
env.reset()
```

```python
n_actions = env.action_space.n
state_dim = env.observation_space.shape[0]
```

```python
cartpole_agent = Agent(state_size=state_dim,
                       action_size=n_actions,
                       hidden_size_1=16,
                       is_stochastic=True)
```

```python
cartpole_agent.watch_me(env, steps=500)
```

## Develop REINFORCE

```python
return_hist = reinforce(cartpole_agent, env)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(len(return_hist)), return_hist, 'b-')
plt.grid(True)
plt.xlabel('Episode')
plt.ylabel('Return')
plt.show()
```
